```python
%%R
filVliq <- c()
filVliq[1] <- getDists(summ,'fit.design','between',c('2779_fil','2779_liq'))
filVliq[2] <- getDists(summ,'fit.design','between',c('3021_fil','3021_liq'))
filVliq[3] <- getDists(summ,'fit.design','between',c('3043_fil','3043_liq'))
filVliq[4] <- getDists(summ,'fit.design','between',c('3061_fil','3061_liq'))
filVliq[5] <- getDists(summ,'fit.design','between',c('3257_fil','3257_liq'))

stoolVfil <- c()
stoolVfil[1] <- getDists(summ,'fit.design','between',c('2779','2779_fil'))
stoolVfil[2] <- getDists(summ,'fit.design','between',c('3021','3021_fil'))
stoolVfil[3] <- getDists(summ,'fit.design','between',c('3043','3043_fil'))
stoolVfil[4] <- getDists(summ,'fit.design','between',c('3061','3061_fil'))
stoolVfil[5] <- getDists(summ,'fit.design','between',c('3257','3257_fil'))

stoolVliq <- c()
stoolVliq[1] <- getDists(summ,'fit.design','between',c('2779','2779_liq'))
stoolVliq[2] <- getDists(summ,'fit.design','between',c('3021','3021_liq'))
stoolVliq[3] <- getDists(summ,'fit.design','between',c('3043','3043_liq'))
stoolVliq[4] <- getDists(summ,'fit.design','between',c('3061','3061_liq'))
stoolVliq[5] <- getDists(summ,'fit.design','between',c('3257','3257_liq'))

stoolVstool <-c()
stoolVstool[1] <- getDists(summ,'fit.design','between',c('2779','3021'))
stoolVstool[2] <- getDists(summ,'fit.design','between',c('2779','3043'))
stoolVstool[3] <- getDists(summ,'fit.design','between',c('2779','3061'))
stoolVstool[4] <- getDists(summ,'fit.design','between',c('2779','3257'))
stoolVstool[5] <- getDists(summ,'fit.design','between',c('3021','3043'))
stoolVstool[6] <- getDists(summ,'fit.design','between',c('3021','3061'))
stoolVstool[7] <- getDists(summ,'fit.design','between',c('3021','3257'))
stoolVstool[8] <- getDists(summ,'fit.design','between',c('3043','3061'))
stoolVstool[9] <- getDists(summ,'fit.design','between',c('3043','3257'))
stoolVstool[10] <- getDists(summ,'fit.design','between',c('3061','3257'))

means<- c(mean(filVliq),mean(stoolVfil),mean(stoolVliq),mean(stoolVstool))
sds<- c(sd(filVliq),sd(stoolVfil),sd(stoolVliq),sd(stoolVstool))

barplot2(means, plot.ci=T, ci.l=means-sds, ci.u=means+sds, ci.lwd=1, ci.col='black', names.arg=c('Liquid vs\nFiltered','Stool vs\nFiltered','Stool vs\nLiquid','Between\nStools'))
```
